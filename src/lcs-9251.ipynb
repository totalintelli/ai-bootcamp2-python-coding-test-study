```python
import sys

# str1 = "ACAYKP"
# str2 = "CAPCAK"
str1 = sys.stdin.readline()
str2 = sys.stdin.readline()
res = list()

for i in range(len(str1)):
    cnt = 0
    for j in range(len(str2)):
        print("str1 = " + str1[i])
        print("str2 = " + str2[j])
        print("-----------------------")
        if (str1[i] == str2[j]) and (i < j) and cnt == 0:
            res.append(str1[i])
            cnt += 1

print(res)
```
